#### Попутно, аналазируя таблицу, видно, что у некоторых переменных только 1 значение в тестовой выборке, думаю, что их можно выбросить, т.к. никакого эффекта на целевую переменную они не окажут.

```python
one_value = []
for i in train:
    if i != 'ID' and i != 'y' and train[i].dtype != 'object':
        if train[i].nunique() == 1:
            one_value.append(i)
print(*one_value)
```

#### Проанализировав, получаем, что таких переменных 12.

```python
train = train.drop(one_value,axis = 1)
train = train.drop('ID',axis = 1)
train
```

```python
d = dict()
for i in train:
    s1 = s0 = 0
    if i not in ['ID','y'] and train[i].dtype == 'int64':
        s0 += train.loc[train[i] == 0, 'y'].mean()
        s1 += train.loc[train[i] == 1, 'y'].mean()
        d[i] = abs(s0-s1)
sns.catplot(x = d.keys(), y = d.values())
```

```python
temp = ({
    'Column_name':[],
    'Abs_value' :[]
})
temp = pd.DataFrame(temp)
for i in train:
    s1 = s0 = 0
    if i not in ['ID','y'] and train[i].dtype == 'int64':
        s0 += train.loc[train[i] == 0, 'y'].mean()
        s1 += train.loc[train[i] == 1, 'y'].mean()
        temp.loc[len(temp.index)] = [i,abs(s0-s1)]
        
temp = temp.sort_values('Abs_value')
plt.figure(figsize=(15,60))
plt.hlines(y=temp['Column_name'], xmin=0, xmax=temp.sort_values('Abs_value')['Abs_value'], color='skyblue')
plt.plot(temp['Abs_value'], temp['Column_name'], "o")
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
plt.margins(0.002,0.002)
```

#### Построили график, где на оси 'y' отложены названия колонок оригинальной таблицы, в качестве 'x' - разность средних значений 'y', с условием когда в столбце переменная принмает значение 1 или 0. Можно заметить, что некоторые столбцы не помогут нам обобщить данные, т.к. разность средних значений близка к 0, однако у некоторых столбцов разность принимает значение больше 10, что поможет нам обобщить данные.

```python
sns.set(rc={'figure.figsize':(10.7,8.27)})
sns.histplot(data = train, x = 'y',kde = True,bins =100)
plt.xlim(train['y'].min(),train['y'].max())
```

#### По графику видно, что в основном целевая переменная принимает значения от 90 до 110. За редким исключением переменная ниже 125.

```python
import catboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
model = catboost.CatBoostRegressor(iterations= 500, learning_rate= 0.005,cat_features=['X0','X1','X2','X3','X4','X5','X6','X8'],depth = 3,eval_metric='RMSE')
x_tr,x_test,y_tr,y_test = train_test_split(train.drop('y',axis = 1),train['y'])
model.fit(x_tr,y_tr)
mean_squared_error(y_test,model.predict(x_test))
```

```python
x_tr1,x_val,y_tr1,y_val = train_test_split(x_tr,y_tr,train_size=0.9)
train_data = catboost.Pool(data = x_tr1,label = y_tr1, cat_features = ['X0','X1','X2','X3','X4','X5','X6','X8'])
eval_data = catboost.Pool(data = x_val,label = y_val, cat_features = ['X0','X1','X2','X3','X4','X5','X6','X8'])

model1 = catboost.CatBoostRegressor(iterations= 1000,depth = 4,  learning_rate= 0.005, eval_metric='RMSE')
model1.fit(train_data,use_best_model=True,eval_set=eval_data)
mean_squared_error(y_test,model1.predict(x_test))
```
